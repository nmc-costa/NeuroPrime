## 4. Experiment 2 Design

Study tested 62 participants grouped in 2 conditions:
- no-priming, N=31, control group (CG)
- priming, N=31, experimental group (EG)

Each participant of CG did a single session of no-priming.

Each participant of EG did a single session of priming.

The session was divided in 6 blocks.

The first and the last block are equal for the two groups, named Block in(Bin) and Block out(Bout):
- REST_eyesclosed+REST_eyesopen+NFT_eyesopen+state self-reports(TMS & POMS)

The 4 blocks between Bin and Bout, B1 to B4, are different for EG and CG:
- All participant went through 2 randomized conditions of eyes closed(ec) and eyes open(eo)
- CG went through a reptition of 4 trials of REST task then NFT task.
- EG went through 2 randomized priming conditions (PRIME task), breathing (BM) and imagery (IM), with 2 trials for each PRIME task then NFT task

### 5.2. (EEG) pipeline & features
#### Input Data 
* Sample rate: 1000Hz

#### Pre-processing
* Data interval Selection: start and end task
* Filter: 1Hz-40Hz (FIR, lfilter)
* Re-referencing: average
* Select channels: Fp1, Fp2, Fz and Pz
* Segmentation into epochs: 1.024s = 1024 points
* Epoch artifact rejection: 
    * STUDY_2: maximum peak to peak of over ±150 μV; a minimum peak acceptance of ±0.5 μV;
    * Result: N_EG=30(S016 missing data); N_CG=31)

#### Processing
* FFT Transformation: n_fft=1024, n_per_seg=1024, n_overlap=256=1024*0.25. 
* PSD channel representation: Pz.
* Get bands: with a standard IAF of 10 Hz: delta(2 to 4 Hz, IAF - 8 to IAF - 6), theta (4 to 8 Hz, IAF - 6 to IAF - 2), lower alpha (8 to 10 Hz, IAF - 2 , IAF), upper alpha (10 to 12 Hz, IAF to IAF+2), overall alpha (8 to 12 Hz, IAF - 2 to IAF + 2), SMR (12-15 Hz, IAF +2 to IAF+5), beta (15 to 35 Hz, IAF+5 to 35) and gamma (35-100 Hz).

#### Features
* Feature ROI: Pz
* Features extracted per subject:

`['theta','alpha','SMR','beta']=EEGBAND`: Bands extracted (#NOTE:  the most important is alpha, as it was the NFT target)

`'epoch_a'`: Bands psd epoch array (~1s)

Three major measures that can be used to detect three types of brain activity changes due to neurofeedback (Dempster & Vernon 2009):

- `['mean','std']`: `Amplitude [V^2/Hz]` in  (Dempster & Vernon 2009). Absolute ampitude of task (band mean PSD Amplitude) and standard deviation. #WARNING- MULTIPLY 10^12 to get [uV^2/Hz] 

- `'tBAT'`: `Percent time` in  (Dempster & Vernon 2009). [normalized measure, 0 to 1] percentage of time in brain activity target => synchronization  = epochs_a(above_threshold)/epochs_a(total)  (Dempster & Vernon 2009). 
    - Percent of time spent with positive feedback, which equals the time spent in the intended brain state. Thus, this is the time spent above or beneath a specific threshold when self-regulating one’s brain activity.
    - #WARNING For desynchronization this measure is not great ( for ec this measure can get to zero) -> #SOLUTION use log10(mean) ; #WARNING #BUG if epochs_above_threshold=0 the algorithm puts the cell empty - the case of S055 alpha_tBAT_task_12 desynchronization (rest_ec Bout) -> #SOLUTION for now it was addressed manually and changed to zero in e2_database_v5.

- `'tonic_increase'` (not a good name, change to `'integrated_change'` as in the thesis): `integrated alpha` in  (Dempster & Vernon 2009). [normalized measure, (0 to 1)* mean] Amplitude(mean)*tBAT (Dempster & Vernon 2009)
    - the percent time spent in the desired brain state * mean level of the amplitude during neurofeedback


Each of these measures seem to reflect different aspects of brain activity, and it might be worthy reporting all measures. Over the course of training it is possible to observe changes in both amplitude(mean) and percent in time(tBAT):
- `'mean'(task absolute amplitude)`, `the learner could exhibit only brief and temporally unstable increases over time`, which nonetheless could be reflected in average amplitude changes (Hardt and Kamiya, 1976). 
- `'tBAT'`,  `the learner could show slight differences within training that are temporally stable` (Dempster & Vernon 2009). 
- `'integrated_change'`, with respect to the comparison of different frequencies, it might be worth calculating these measures in both the absolute amplitude values as well as changes in percent, since amplitude scales with frequency. Integrated change fails to provide information about the two measures independently. Thus, it would seem far more informative to discuss amplitude and percent time separately rather than combining them to form one integrated measure (Dempster & Vernon 2009).

Other features: 
- `'threshold'`: offline threshold used in each task = Baseline REST task mean (Block in): ec, “rest_ec_1”;eo “rest_eo_2” (Bin);  #WARNING online training threshold was adaptative to the subject.

- `'ratio_power_threshold'`: [normalized measure, -1 to 1] Amplitude/baseline (amplitude for groups - with baseline normalization using initial threshold) => #WARNING for tasks_ec the initial threshold is rest_ec mean; for tasks_eo the baseline is rest_eo mean; #WARNING this means that task_1 and task_2 is allways 1;  good for both sync/desync

#### TIME DOMAIN FEATURES
Mostly used on long term recordings (24h) but some studies use some of those features on short term recordings, from 1 to 5 minutes window.

- mean_nni: The mean of RR-intervals.
- sdnn : The standard deviation of the time interval between successive normal heart beats (i.e. the RR-intervals).
- sdsd: The standard deviation of differences between adjacent RR-intervals
- rmssd: The square root of the mean of the sum of the squares of differences between adjacent NN-intervals. Reflects high frequency (fast or parasympathetic) influences on hrV (i.e., those influencing larger changes from one beat to the next).
- median_nni: Median Absolute values of the successive differences between the RR-intervals.
- nni_50: Number of interval differences of successive RR-intervals greater than 50 ms.
- pnni_50: The proportion derived by dividing nni_50 (The number of interval differences of successive RR-intervals greater than 50 ms) by the total number of RR-intervals.
- nni_20: Number of interval differences of successive RR-intervals greater than 20 ms.
- pnni_20: The proportion derived by dividing nni_20 (The number of interval differences of successive RR-intervals greater than 20 ms) by the total number of RR-intervals.
- range_nni: difference between the maximum and minimum nn_interval.
- cvsd: Coefficient of variation of successive differences equal to the rmssd divided by mean_nni.
- cvnni: Coefficient of variation equal to the ratio of sdnn divided by mean_nni.
- mean_hr: The mean Heart Rate.
- max_hr: Max heart rate.
- min_hr: Min heart rate.
- std_hr: Standard deviation of heart rate.

References:

[1]	Heart rate variability - Standards of measurement, physiological interpretation, and clinical use, Task Force of The European Society of Cardiology and The North American Society of Pacing and Electrophysiology, 1996


#### FREQUENCY DOMAIN FEATURES
To our knowledge, you might use this function on short term recordings, from 2 to 5 minutes window.

- total_power : Total power density spectral
- vlf : variance ( = power ) in HRV in the Very low Frequency (.003 to .04 Hz by default). Reflect an intrinsic rhythm produced by the heart which is modulated primarily by sympathetic activity.
- lf : variance ( = power ) in HRV in the low Frequency (.04 to .15 Hz). Reflects a mixture of sympathetic and parasympathetic activity, but in long-term recordings, it reflects sympathetic activity and can be reduced by the beta-adrenergic antagonist propanolol.
- hf: variance ( = power ) in HRV in the High Frequency (.15 to .40 Hz by default). Reflects fast changes in beat-to-beat variability due to parasympathetic (vagal) activity. Sometimes called the respiratory band because it corresponds to HRV changes related to the respiratory cycle and can be increased by slow, deep breathing (about 6 or 7 breaths per minute) and decreased by anticholinergic drugs or vagal blockade.
- lf_hf_ratio : lf/hf ratio is sometimes used by some investigators as a quantitative mirror of the sympatho/vagal balance.
- lfnu : normalized lf power.
- hfnu : normalized hf power.

References:

### 6.2. [Theory] Measures assessing framework successful control 

Theoretically, an NFT framework implies that any observable measure of brain activity can be extracted and tested for volitional control. Nonetheless, what constitutes successful control, and how to quantify it? In the engineering sense, successful control can be viewed as enhancing the signal-to-noise ratio of a parameter relative to a control condition, a reference condition (e.g., resting-state, sham, or sensory stimulation without control), which could be administered sequentially or interspersed randomly in the experiment (Tomas Ros et al., 2014). Dempster and Vernon (D Vernon et al., 2009) suggested three major measures that can be used to assess feature changes of brain activity during NF:
1. Changes in absolute values of frequency band amplitude/power (e.g. (Angelakis et al., 2007)).
2. Percentage of time in brain activity target (tBAT), basically the percent time spent in the desired brain state. Thus, this is the time spent above or beneath a specific threshold when self-regulating one’s brain activity.
3. Tonic increase, multiplication of the absolute amplitude (1.) with tBAT (2.).

However, there are more oscillatory parameters reported to be volitionally controlled including phase-locking (Brunner et al., 2006) and nonlinear fractal dimension (Wang et al., 2011).
Based on these measures, one can assess feature changes in four different ways (S Enriquez-Geppert & Huster, 2017) : (1) within sessions, for instance by comparing the beginning of each session with the end of a training session (sessions can be arbitrarily divided into blocks or segments for statistical analyses); (2) possible changes in these measures within a session observed as a difference relative to a baseline measurement, where participants neither try to control their brain activity nor receive any feedback; as (3) changes observed from session-to-session; and accordingly also as (4) changes across sessions relative to a baseline measure.

Afterwards, `hypothesis testing` may be applied to test whether during NFT there is a significant difference in the mean together with a reduction (or no change) in the variability of the controlled signal (between the different conditions). With respect to existing methods in the literature, this approach is technically equivalent to an analysis of variance (ANOVA) or a Student’s t-test, which similarly account for a variable’s mean and variance. If multiple confounding variables are involved, it might then be appropriate to use a multivariate analysis of covariance (MANCOVA) (Tomas Ros et al., 2014).
